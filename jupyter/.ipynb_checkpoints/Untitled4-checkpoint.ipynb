```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response = requests.get("https://punchng.com/topics/news/")
html_data = soup(response.text)
```

```python
news_cards = html_data.find_all("div")
```

```python
for news_card in news_cards:
    print(news_card.text)
```
