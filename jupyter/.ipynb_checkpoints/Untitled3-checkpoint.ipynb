```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response = requests.get("https://punchng.com/topics/news/")
html_data = soup(response.text)     #parse html data into bs4
```

```python
news_card = html_data.find_all("div")    #find all div tags. including the unwanted cards
```

```python
for news_card in news_card:
    print(news_card.text)
```
