```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response = requests.get("https://punchng.com/topics/news/")
html_data = soup(response.text)
```

```python
news_cards = html_data.find_all("div")
```

```python
for news_card in news_cards:
    print(news_card.text)
```

```python
news_cards = html_data.find_all("div", {"class", "grid-item"})
```

```python
news_cards
```

```python
for news_card in news_cards:
    link_tag = news_card.find("a")
    
    print(link_tag.attrs["href"])
    print(news_card.text.strip(), end ="n\n\n")
```

```python
for news_card in news_cards:
    link_tag = news_card.find("a")
    link = link_tag.attrs["href"]
    
    link_response = requests.get(link)
    link_html_data = soup(link_response.text)
    news = link_html_data.find("div", {"class":"entry-content"})
    
    print(link_tag.attrs["href"])
    print(news_card.text.strip(), end="n\n")
    print(news.text.strip(), end="n\n")
```

```python
url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 2):
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data = soup(response.text)
    
    news_cards = html_data.find_all("div", {"class", "grid-item"})
    
    for news_card in news_cards:  #now more conscise
        link_tag = news_card.find("a")
        link = link_tag.attrs['href']
        
        link_response = requests.get(link)
        link_html_data = soup(link_response.text)
        news = link_html_data.find("div", {"class":"entry-content"})
    
        print(link_tag.attrs["href"])
        print(news_card.text.strip(), end="n\n")
        print(news.text.strip(), end="n\n")
        
        print(f"----------PAGE {page_number}------------")
```

```python
url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 2):
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data = soup(response.text)
    
    news_cards = html_data.find_all("div", {"class", "grid-item"})
    
    for news_card in news_cards:  #now more conscise
        link_tag = news_card.find("a")
        link = link_tag.attrs['href']
        
        link_response = requests.get(link)
        link_html_data = soup(link_response.text)
        news = link_html_data.find("div", {"class":"entry-content"})
        
        file_name = news_card.text.strip()
        mode = "w"
        file = open(file_name, mode)
        file.write(link_tag.attrs["href"])
        file.write(news.text.strip())
        
        print(file)
        file.close
        
        print(link_tag.attrs["href"])
        print(news_card.text.strip(), end="n\n")
        print(news.text.strip(), end="n\n")
        
        print(f"----------PAGE {page_number}------------")
```
